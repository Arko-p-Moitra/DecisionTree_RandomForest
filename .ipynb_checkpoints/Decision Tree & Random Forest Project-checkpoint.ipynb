```python
#Importing libs here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing csv here

loanCSV=pd.read_csv("loan_data.csv")
```

```python
loanCSV.head()
```

```python
loanCSV.info()
```

```python
loanCSV.describe()
```

# Data analysis and visualization

```python
plt.figure(figsize=(10,6))
loanCSV[loanCSV['credit.policy']==1]['fico'].hist(alpha=0.5,color='blue',
                                              bins=30,label='Credit.Policy=1')
loanCSV[loanCSV['credit.policy']==0]['fico'].hist(alpha=0.5,color='red',
                                              bins=30,label='Credit.Policy=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(10,6))
loanCSV[loanCSV['not.fully.paid']==1]['fico'].hist(alpha=0.5,color='blue',
                                              bins=30,label='not.fully.paid=1')
loanCSV[loanCSV['not.fully.paid']==0]['fico'].hist(alpha=0.5,color='red',
                                              bins=30,label='not.fully.paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loanCSV,palette='Set1')
```

```python
sns.jointplot(x='fico',y='int.rate',data=loanCSV,color='purple')
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loanCSV,hue='credit.policy',
           col='not.fully.paid',palette='Set1')
```
